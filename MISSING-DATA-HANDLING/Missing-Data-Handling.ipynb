```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('P_D.csv')
ds1=pd.read_csv('P_D.csv')
ds11=pd.read_csv('P_D.csv')
```

```python
ds.head()
```

```python
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds=ds.dropna(how='any')
ds.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.histplot(ds1.salary, kde=True)
```

```python
ds1['salary'].fillna(ds1['salary'].median(), inplace=True)
```

```python
ds1.isnull().sum()
```

```python
fig , ax = plt.subplots(figsize=(8,8))
sns.histplot(ds1.salary,kde=True)
```

```python
ds11['salary'].fillna(ds11['salary'].mode(),inplace=True)
```

```python
fig , ax = plt.subplots(figsize=(8,8))
sns.histplot(ds11.salary,kde=True)
```
